```python
#Creating a copy of the income column so we don't edit the values 
income_copy = df['income'].copy() 

#Creating the z-score values for each 

income_std = (income_copy - income_copy.mean()) / income_copy.std()
print(income_std.mean()) 
print(income_std.std())
income_std.head(5)
```

6.  The mean of the normalized income attribute is about -2.514. The standard deviation of the normalized income attribute is about 1.000. This is because once we normalized the attribute we can tell truly how far each observation is from the mean of the original values by standard deviations. 

```python
#Storing in age and not changing the values 
age_copy = df['age'].copy()

age_bins = pd.qcut(age_copy, [0,.33,.67,1],  labels=['young','mid-age','old'])
age_bins.value_counts()
```

```python
#8: 
bank_norm = df.copy()

bank_norm = pd.get_dummies(bank_norm)

#min-max normalization 
bank_norm = (bank_norm - bank_norm.min()) / (bank_norm.max() - bank_norm.min())

bank_norm.head(5)

```

```python
print(bank_norm['income'].mean()) 
print(bank_norm['income'].std())
```

8. The mean of income after min-max normalization is .38732645 and standard deviation is .22196114. These values are different than z-score normalization because we are keeping the values in the range of 0-1 as well as doing a different form of normalization. 

```python
#9: 
bank_standard = bank_norm.iloc[:, 0:]  
bank_standard = pd.get_dummies(bank_norm) 
bank_standard.head(1)
```
